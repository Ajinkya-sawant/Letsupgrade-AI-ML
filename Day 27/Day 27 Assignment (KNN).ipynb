```python
import pandas as pd
from sklearn import preprocessing
```

```python
df=pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.info()
```

## *Encoding of categorical variables*

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
df['Sex']=encoder.fit_transform(df['Sex'])
df['Embarked']=encoder.fit_transform(df['Embarked'])
```

```python
df.head()
```

## *Check for null values*

```python
df.isna().sum()
```

## *Dropping of irrelevant features*

```python
df=df.drop(['PassengerId','Cabin','Ticket','Name'],axis=1)
df.head()
```

## *Splitting DataSet*

```python
y=df['Pclass']
y.head()
```

```python
x=df.drop(['Pclass'],axis=1)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
from sklearn import neighbors
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
    # Train model
def KNN(n):
    for i in range(1,n):
        knn = neighbors.KNeighborsClassifier(n_neighbors=i)
        print("\nvalue of n_neighbors =",i)
        knn.fit(x_train, y_train)
        y_preds=knn.predict(x_test)
        print("\n Confusion Matrix:\n",confusion_matrix(y_test,y_preds))
        print("\n Accuracy Score \n",accuracy_score(y_test,y_preds))
        print("-----------------------------------------------------------------------------------------")
```

```python
KNN(268)
```

## *Conclusion*

**From above it is clear that accuracy score is high 89%, when n_neighbors value is 1 and out of 267 records 238 records are correctly classified**
